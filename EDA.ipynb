```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['weight'], bins=50, kde=True)
plt.title('Distribution of Weight (Stock)')
plt.show()
```

```python
top_brands = df.groupby('brandname')['weight'].sum().sort_values(ascending=False)
top_brands.plot(kind='bar', figsize=(10,6), title='Total Stock per Brand')
plt.ylabel('Total Weight')
plt.tight_layout()
plt.show()
```

```python
product_stock = df.groupby('productname')['weight'].sum().sort_values(ascending=False).head(10)
product_stock.plot(kind='barh', title='Top 10 Products by Stock')
plt.xlabel('Total Stock (Weight)')
plt.tight_layout()
plt.show()

```
